# Section 1: Distribution of Vehicle Classes (VClass)
# This bar plot shows the count of each vehicle class (VClass) in the dataset.


```python
plt.figure(figsize=(10, 6))
sns.countplot(y='VClass', data=data, order=data['VClass'].value_counts().index)
plt.title('Distribution of Vehicle Classes')
plt.xlabel('Count')
plt.ylabel('Vehicle Class')
plt.show()
```

# Section 2: Fuel Type Distribution (Pie Chart)
# A pie chart visualizing the proportion of each fuel type (fuelType) in the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the figure with a larger size for better visibility
plt.figure(figsize=(10, 8))

# Calculate fuel type data counts
fuel_type_data = data['fuelType'].value_counts()

# Define colors for the pie chart
colors = sns.color_palette('pastel', len(fuel_type_data))  # Ensure we have enough colors for each category

# Draw the pie chart without labels and percentages inside the chart
plt.pie(fuel_type_data,
        startangle=140,
        colors=colors,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'})

# Create the table to show in the upper-right corner with the fuel types, percentages, and color samples
table_data = pd.DataFrame({
    'Fuel Type': fuel_type_data.index,
    'Percentage': (fuel_type_data / fuel_type_data.sum() * 100).round(1),
    'Color': colors  # Add the colors corresponding to each fuel type
})

# Create table content with color samples
table_content = [[f'{fuel_type}', f'{percentage}%', ''] for fuel_type, percentage in zip(table_data['Fuel Type'], table_data['Percentage'])]

# Create the table and display it in the upper-right corner
the_table = plt.table(cellText=table_content,
                      colLabels=['Fuel Type', 'Percentage (%)', 'Color'],
                      cellLoc='center', colLoc='center', loc='upper right', bbox=[1.15, 0.5, 0.25, 0.25])

# Adjust the font size of the table to be proportional to the figure size
the_table.auto_set_font_size(False)
the_table.set_fontsize(12)  # Adjust the font size for better readability
the_table.scale(1.2, 1.2)  # Slightly scale up the table for better clarity

# Add colors to the cells in the "Color" column
for i, color in enumerate(colors):
    the_table[(i + 1, 2)].set_facecolor(color)  # Set the background color of the color cells

# Set the border and font weight for better presentation
for key, cell in the_table.get_celld().items():
    cell.set_linewidth(1.5)  # Thicker borders for clarity
    if key[0] == 0:  # Header row
        cell.set_fontsize(13)
        cell.set_text_props(weight='bold')

# Add the chart title
plt.title('Fuel Type Distribution', fontsize=16)

# Apply tight layout to ensure everything fits well
plt.tight_layout()

# Show the final plot
plt.show()
```

# Section 3: City vs Highway MPG Comparison (Scatter Plot)
# A scatter plot that compares City MPG vs Highway MPG, colored by vehicle class.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='city', y='highway', data=data, hue='VClass', palette='viridis', alpha=0.7)
plt.title('City vs Highway MPG')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.show()
```

# Section 4: CO2 Emissions by Vehicle Class (Box Plot)
# A box plot that shows the distribution of CO2 emissions for each vehicle class.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='VClass', y='co2', data=data, palette='Set2')
plt.xticks(rotation=90)
plt.title('CO2 Emissions by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```

# Section 5: Fuel Economy Score Distribution (Histogram)
# A histogram displaying the distribution of fuel economy scores across all vehicles.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['feScore'], bins=10, kde=True, color='blue')
plt.title('Fuel Economy Score Distribution')
plt.xlabel('Fuel Economy Score')
plt.ylabel('Frequency')
plt.show()
```

# Section 6: Cylinders vs Displacement (Scatter Plot)
# This scatter plot shows the relationship between the number of cylinders and engine displacement, colored by fuel type.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cylinders', y='displ', data=data, hue='fuelType', palette='muted', alpha=0.7)
plt.title('Cylinders vs Displacement')
plt.xlabel('Number of Cylinders')
plt.ylabel('Engine Displacement (Liters)')
plt.show()
```

# Section 7: Average CO2 Emissions by Transmission Type (Bar Plot)
# A bar plot showing the average CO2 emissions for each type of vehicle transmission.
# Dropping missing values in CO2 and transmission columns for a cleaner dataset.

```python
data_cleaned = data.dropna(subset=['co2', 'trans'])
plt.figure(figsize=(10, 6))
sns.barplot(x='trans', y='co2', data=data_cleaned, ci=None, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Average CO2 Emissions by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.show()
```

# Section 8: Fuel Economy (Combined) Across Car Makes (Violin Plot)
# A violin plot visualizing the combined fuel economy for different car manufacturers.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='make', y='comb', data=data_cleaned, palette='Set3')
plt.xticks(rotation=90)
plt.title('Fuel Economy (Combined) Across Car Makes')
plt.xlabel('Car Make')
plt.ylabel('Combined Fuel Economy (MPG)')
plt.show()
```

# Section 9: Average Fuel Efficiency by Drive Type (Bar Plot)
# A bar plot showing the average fuel efficiency (combined MPG) for different drive types.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='drive', y='comb', data=data_cleaned, estimator=np.mean, ci=None, palette='magma')
plt.title('Average Fuel Efficiency by Drive Type')
plt.xlabel('Drive Type')
plt.ylabel('Average Combined MPG')
plt.show()
```

# Section 10: Yearly Trend of CO2 Emissions (Line Plot)
# A line plot showing the trend of average CO2 emissions over the years.

```python
yearly_co2 = data_cleaned.groupby('year')['co2'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='co2', data=yearly_co2, marker='o', color='green')
plt.title('Yearly Trend of CO2 Emissions')
plt.xlabel('Year')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.show()
```
